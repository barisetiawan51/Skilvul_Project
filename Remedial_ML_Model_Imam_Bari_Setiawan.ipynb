# Problem Definition
## Latar Belakang
Migrain merupakan salah satu masalah kesehatan yang umum dijumpai di berbagai belahan dunia dan mempengaruhi jutaan orang di berbagai kelompok usia dan latar belakang. Kompleksitas dalam diagnosis dan pengelolaannya masih menjadi permasalahan umum yang dihadapi. Meskipun telah ada kemajuan yang signifikan dalam memahami penyebab migrain, diagnosis seringkali mmengandalkan pada gejala yang dilaporkan oleh pasien, dapat bervariasi secara signifikan dari satu individu ke individu lainnya. Ini menjadi tantangan dalam menetapkan diagnosis yang akurat dan menyesuaikan pengobatan secara tepat.

Secara spesifik, penelitian mengenai migrain telah menunjukkan bahwa faktor-faktor seperti usia, jenis kelamin dan faktor lingkungan dapat mempengaruhi frekuensi dan intensitas serangan migrain. Namun, mengidentifikasi pola-pola ini secara manual dari data yang besar dan kompleks menjadi tantangan tersendiri.  Maka dibutuhkan machine learning yang dapat memberikan kontribusi yang berharga dan bernilai.

## Data yang akan dipakai
1. Nama dataset adalah Data for Migraine Classification.
2. Sumber data diperoleh dari Kaggle dengan alamat URL sebagai berikut: [Dataset Migrain](https://www.kaggle.com/datasets/weinoose/migraine-classification/data)
3. Deskripsi atau detail dapat dijabarkan dibawah ini:
    * Age (Usia): Ini adalah kolom yang mencatat usia pasien yang mengalami migrain. Informasi usia ini penting karena migrain bisa mempengaruhi orang dari segala rentang usia, meskipun kecenderungannya berkurang seiring bertambahnya usia. Pada dataset ini rentang usia dimulai dari usia 15 tahun sampai 77 tahun.

    * Duration (Durasi): Kolom ini mencatat durasi serangan migrain, yaitu berapa lama pasien mengalami migrain. Durasi tersebut dihitung dalam satuan hari dengan retang waktu 1 sampai 3, memberikan gambaran tentang seberapa lama pasien harus menghadapi rasa sakit dan ketidaknyamanan. Hal tersebut mengacu pada berapa lama serangan migrain berlangsung pada kapan terakhir migrain dialami oleh pasien.

    * Frequency (Frekuensi): Kolom ni mencatat seberapa sering pasien mengalami migrain per bulan. Rentang waktu yang diberikan pada frekuensi ini berkisar 1-8. Informasi ini memberikan pemahaman tentang seberapa sering pasien harus menanggung beban migrain, yang bisa berkisar dari sporadis hingga sangat sering.

    * Location (Lokasi): Kolom ini mengidentifikasi lokasi sakit kepala pada pasien. Lokasi nyeri kepala yang dimaksudkan yaitu jika terdapat pada satu sisi (unilateral), kedua sisi (bilateral), atau tidak ada nyeri kepala sama sekali (None). Nilai-nilai yang diberikan adalah 0 (Tidak ada), 1 (Unilateral), atau 2 (Bilateral).

    * Character (Karakteristik): Kolom Ini adalah deskripsi dari karakteristik rasa sakit migrain yang dialami oleh pasien. karakteristik nyeri kepala dibedakan menjadi, apakah itu berdenyut (Throbbing), konstan (Constant), atau tidak ada karakteristik tertentu (None). Nilai-nilai yang diberikan adalah 0 (Tidak ada), 1 (Throbbing), atau 2 (Constant).

    * Intensity (Intensitas): Kolom Ini mencatat tingkat keparahan rasa sakit migrain yang dirasakan oleh pasien. Intensitas ini diukur dengan kategori berikut, ringan (Mild), sedang (Medium), atau parah (Severe). Nilai-nilai yang diberikan adalah 0 (Tidak ada), 1 (Mild), 2 (Medium), atau 3 (Severe).

    * Nausea (Mual): Ini adalah variabel biner yang menunjukkan apakah pasien mengalami mual selama migrain (1 untuk ya, 0 untuk tidak).

    * Vomit (Muntah): Ini adalah variabel biner yang menunjukkan apakah pasien mengalami muntah selama migrain (1 untuk ya, 0 untuk tidak).

    * Phonophobia (Fonofobia): Ini adalah variabel biner yang menunjukkan apakah pasien memiliki sensitivitas terhadap suara yang meningkat sebagai gejala migrain (1 untuk ya, 0 untuk tidak).

    * Photophobia (Fotofobia): Ini adalah variabel biner yang menunjukkan apakah pasien memiliki sensitivitas terhadap cahaya yang meningkat sebagai gejala migrain (1 untuk ya, 0 untuk tidak).

    * Visual (Visual): Kemungkinan merujuk pada gejala visual yang dialami selama migrain, seperti aurora visual.

    * Sensory (Sensorik): Kemungkinan merujuk pada gejala sensorik yang dialami selama migrain, seperti kesemutan atau kebasan.

    * Dysphasia (Disfagia): Kemungkinan merujuk pada kesulitan dalam menelan yang dialami selama migrain.

    * Dysarthria (Disartria): Kemungkinan merujuk pada gangguan bicara yang dialami selama migrain.

    * Vertigo (Vertigo): Kemungkinan merujuk pada rasa pusing atau perasaan berputar yang dialami selama migrain.

    * Tinnitus (Tinnitus): Kemungkinan merujuk pada gejala berdengung atau berdering di telinga yang terjadi selama migrain.

    * Hypoacusis (Hipoakusis): Kemungkinan merujuk pada penurunan pendengaran yang terjadi selama migrain.

    * Diplopia (Diplopia): Kemungkinan merujuk pada penglihatan ganda yang dialami selama migrain.

    * Defect (Kekurangan): Kemungkinan merujuk pada kekurangan atau gangguan sensorik yang terjadi selama migrain.

    * Ataxia (Ataksia): Kemungkinan merujuk pada masalah koordinasi gerakan yang terjadi selama migrain.

    * Conscience (Kesadaran): Kemungkinan merujuk pada tingkat kesadaran subjek selama migrain.

    * Paresthesia (Parestesia): Kemungkinan merujuk pada sensasi kesemutan, terbakar, atau mati rasa yang terjadi selama migrain.

    * DPF: Latar belakang keluarga (Yes) atau tidak (Not) yang mungkin memengaruhi risiko atau predisposisi seseorang terhadap migrain.

    * Type (Tipe): Diagnosis tipe migrain yang dialami oleh pasien. Ini mencakup berbagai jenis migrain seperti migrain dengan aura, migrain tanpa aura, migrain dengan aura tanpa migrain, migrain hemiplegik familial, migrain hemiplegik sporadik, aura tipe basilar, atau jenis lainnya. Angka dalam kurung menunjukkan jumlah kasus untuk masing-masing tipe migrain.

### Resampling Data

Resampling data dilakukan untuk mengubah jumlah sampel atau frekuensi data dengan tujuan agar dataset ini memiliki keseimbangan kelas. Dengan melakukan oversampling pada kelas minoritas untuk memperbaiki ketidakseimbangan agar kinerja pada kelas minoritas meningkat.

Resampling data ini menambahkan sejumlah sampel sebanyak 200 pada setiap sampel Type. Dengan hal tesebut sampel akan bertambah sebanyak 7x200.

````python
data_1 = data_migrain[data_migrain['Type'] == 'Typical aura with migraine']
data_2 = data_migrain[data_migrain['Type'] == 'Migraine without aura']
data_3 = data_migrain[data_migrain['Type'] == 'Familial hemiplegic migraine']
data_4 = data_migrain[data_migrain['Type'] == 'Typical aura without migraine']
data_5 = data_migrain[data_migrain['Type'] == 'Basilar-type aura']
data_6 = data_migrain[data_migrain['Type'] == 'Other']
data_7 = data_migrain[data_migrain['Type'] == 'Sporadic hemiplegic migraine']
````

````python
from sklearn.utils import resample

data_1_resample = resample(data_1, n_samples=200, random_state=123, replace=True)
data_2_resample = resample(data_2, n_samples=200, random_state=123, replace=True)
data_3_resample = resample(data_3, n_samples=200, random_state=123, replace=True)
data_4_resample = resample(data_4, n_samples=200, random_state=123, replace=True)
data_5_resample = resample(data_5, n_samples=200, random_state=123, replace=True)
data_6_resample = resample(data_6, n_samples=200, random_state=123, replace=True)
data_7_resample = resample(data_7, n_samples=200, random_state=123, replace=True)

data_migrain = pd.concat([data_1, data_2_resample, data_3_resample, data_4_resample,
                          data_5_resample, data_6_resample, data_7_resample])

data_migrain.head()
````

# Model Training | Pelatihan Model

## Random Forest

````python
# memisahkan atribut dan label
X = data_migrain[['Age', 'Duration', 'Frequency', 'Location', 'Character', 'Intensity', 'Nausea', 'Vomit',
                  'Phonophobia', 'Photophobia','Visual', 'Sensory', 'Dysphasia', 'Dysarthria', 'Vertigo',
                  'Tinnitus', 'Hypoacusis', 'Diplopia', 'Defect', 'Conscience', 'Paresthesia', 'DPF']]
y = data_migrain['Type']

# Bagi data menjadi data latih dan data uji
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.1, random_state=123)

# Inisialisasi model Random Forest dengan hyperparameter
random_forest = RandomForestClassifier(max_depth = None, min_samples_split = 2, min_samples_leaf = 1, n_estimators = 300, random_state=100, class_weight="balanced")

# Latih model Random Forest menggunakan data latih
random_forest.fit(X_train, y_train)
````

Implementasi model Random Forest ini diinisialisasi dengan hyperparameter tertentu yang mencakup:
1. 'max_depth' (kedalaman maksimum pohon) yang bernilai None berarti tidak ada batasan nilai maksimum dari setiap kedalaman.
2. 'min_samples_split' (jumlah sampel minimum yang diperlukan untuk membagi simpul) yang bernilai 2.
3. 'min_samples_leaf' (jumlah sampel minimum yang diperlukan untuk menjadi simpul daun) yang bernilai 1.
4. 'n_estimators' (jumlah pohon dalam hutan) yang bernilai 300.
5. 'random_state' (penentu seed untuk menghasilkan hasil yang dapat direproduksi) yang bernilai 100.
6. 'class_weight' (penanganan ketidakseimbangan kelas dengan memberikan bobot kelas) yang bernilai balanced. Dengan hal ini akan meminimalisir overfitting pada model.

````python
# Lakukan prediksi pada data uji
RF_pred = random_forest.predict(X_test)

# Hitung akurasi prediksi
RF_score = round(accuracy_score(y_test, RF_pred), 2)
print("Akurasi model Random Forest:", RF_score)
````

````python
# Menghitung matriks kebingungan
conf_matrix = confusion_matrix(y_test, RF_pred)

# Visualisasi matriks kebingungan
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=random_forest.classes_, yticklabels=random_forest.classes_)
plt.show()
````

Confusion Matriks merupakan tabel yang digunakan untuk mengevaluasi kinerja dari model klasifikasi. Diagonal utama dari matriks menunjukkan jumlah sampel yang diklasifikasikan dengan benar untuk setiap kelas, sedangkan sel-sel di luar diagonal utama menunjukkan jumlah sampel yang salah diklasifikasikan.

Dengan menggunakan confusion_matrix(y_test, RF_pred), dapat digunakan untuk menghitung matriks kebingungan dengan membandingkan label yang sebenarnya (y_test) dengan label yang diprediksi oleh model Random Forest (RF_pred). Kemudian, matriks kebingungan tersebut divisualisasikan menggunakan sns.heatmap() untuk membuat heatmap yang menunjukkan distribusi hasil klasifikasi secara visual. Anotasi angka di dalam sel heatmap menunjukkan jumlah sampel yang terkandung di setiap sel matriks kebingungan. Heatmap ini memberikan gambaran yang jelas tentang seberapa baik model melakukan klasifikasi untuk setiap kelas yang ada.

Maksud dari angka 1-7 merupakan hasil one hot encoding yang telah dilakukan sebelumnya.

## Logistic Regression

````python
# memisahkan atribut dan label
X = data_migrain[['Age', 'Duration', 'Frequency', 'Location', 'Character', 'Intensity', 'Nausea', 'Vomit',
                  'Phonophobia', 'Photophobia','Visual', 'Sensory', 'Dysphasia', 'Dysarthria', 'Vertigo',
                  'Tinnitus', 'Hypoacusis', 'Diplopia', 'Defect', 'Conscience', 'Paresthesia', 'DPF']]
y = data_migrain['Type']

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)

# Inisialisasi model Regresi Logistik dengan hyperparameter
logistic_regression = LogisticRegression(max_iter=100, random_state=100, n_jobs=20, C=10, solver='newton-cg', class_weight=None)

# Latih model Regresi Logistik menggunakan data latih
logistic_regression.fit(X_train, y_train)
````

Pada pelatihan model ini digunakan hyperparameter yang diinisialisasi oleh model Logistic Regresion yang mana bahwa:
1. 'max_iter' yang bernilai 100 merupakan jumlah maksimum iterasi yang digunakan dalam proses pelatihan model (nilai defaultnya adalah 100).
2. 'random_state' yang bernilai 100 merupakan seed atau nilai acak yang digunakan untuk mengontrol randomization, hal tersebut memastikan bahwa hasil yang dihasilkan dapat direproduksi dengan menjalankan nilai yang sama pada seed yang sama.
3. 'n_jobs' yang bernilai 20 menandakan jumlah pekerjaan pararel yang akan dijalankan selama pelatihan model.
4. 'C' yang bernilai 10 merupakan parameter regularisasi yang mengontrol kekuatan regularisasi dalam model (nilai yang lebih kecil menunjukkan reguralisasi yang lebih kuat).
5. 'solver' yang bernilai newton-cg merupakan algoritma yang digunakan untuk menyelesaikan masalah optimasi.
6. 'class_weight' merupakan pengaturan bobot kelas untuk menangani ketidakseimbangan kelas (nilai None menunjukkan bahwa semua kelas memiliki bobot yang sama).

````python
# Lakukan prediksi pada data uji
LR_pred = logistic_regression.predict(X_test)

# Hitung akurasi prediksi
LR_score = round(accuracy_score(y_test, LR_pred), 2)
print("Akurasi model Regresi Logistik:", LR_score)
````

````python
# Menghitung matriks kebingungan
conf_matrix = confusion_matrix(y_test, LR_pred)

# Visualisasi matriks kebingungan
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=logistic_regression.classes_, yticklabels=logistic_regression.classes_)
plt.show()
````

##  K-Nearest Neighbors (KNN)

````python
# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)

# Inisialisasi model KNN
knn = KNeighborsClassifier(n_neighbors = 5, weights='distance', algorithm='kd_tree', p=1, leaf_size = 10)

# Latih model KNN menggunakan data latih
knn.fit(X_train, y_train)
````

Pada pelatihan model ini diinisialisasikan hyperparameter yang ditentukan dengan:
1. 'n_neighbors' yang bernilai 5 merupakan jumlah tetangga yang akan dipertimbangkan saat melakukan prediksi.
2. 'weights' yang bernilai distance merupakan bobot invers jarak yang akan diterapakan untuk menimbang jarak antar tetangga.
3. 'algorithm' yang bernilai kd_tree merupakan algoritma yang akan digunakan untuk menghitung tetangga terdekat.
4. 'p' yang bernilai 1 mengindikasikan penggunaan jarak manhattan yang merupakan parameter yang menentukan jenis jarak yang digunakan.
5. 'leaf_size' dengan nilai 10 merupakan ukuran daun pada kd_tree untuk pengindeksan.

````python
# Lakukan prediksi pada data uji
KNN_pred = knn.predict(X_test)

# Hitung akurasi prediksi
KNN_score = round(accuracy_score(y_test, KNN_pred), 2)
print("Akurasi model KNN:", KNN_score)
````

````python
# Menghitung matriks kebingungan
conf_matrix = confusion_matrix(y_test, KNN_pred)

# Visualisasi matriks kebingungan
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.show()
````

## Decision Tree

1. Precision adalah proporsi dari positif yang diprediksi dengan benar dari semua instance yang diprediksi positif.
2. Recall adalah proporsi dari positif yang diprediksi dengan benar dari semua instance yang sebenarnya positif.

Support adalah jumlah instance dalam setiap kelas yang terdapat dalam data uji.
4. Accuracy adalah proporsi dari instance yang diklasifikasikan dengan benar dari total instance dalam data uji.
5. Rata-rata makro mengambil rata-rata dari metrik untuk setiap kelas tanpa memperhitungkan distribusi kelas.
6. Rata-rata terponderasi memberikan bobot yang lebih besar pada kelas yang lebih banyak.
